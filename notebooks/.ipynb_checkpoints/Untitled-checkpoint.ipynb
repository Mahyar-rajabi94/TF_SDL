```python
from thermo import Chemical

# Define the chemical (e.g., ethanol)
chemical = Chemical('ethanol')

# Get heat capacity in J/(mol·K)
Cp_molar = chemical.Cp  # Heat capacity at constant pressure in J/(mol·K)

# Get molar mass in g/mol and convert to kg/mol
molar_mass = chemical.MW / 1000  # Convert g/mol to kg/mol

# Convert to J/(kg·K)
Cp_mass = Cp_molar / molar_mass
print(f"Heat Capacity of Ethanol: {Cp_mass:.2f} J/(kg·K)")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
import re

# Define the base URL and parameters
base_url = "https://pubchem.ncbi.nlm.nih.gov/compound/{compound_name}#section={section_name}&fullscreen=true"


def get_temperature_info(compound_name, section_name):
    # Construct the URL
    url = base_url.format(compound_name=compound_name, section_name=section_name)
    
    # Initialize Selenium WebDriver
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # Run in headless mode (no browser window)
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    
    # Load the page
    driver.get(url)
    time.sleep(1)  # Wait for JavaScript content to load (adjust if necessary)

    # Parse the fully rendered page with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # Close the driver
    driver.quit()
    
    # Search for the first occurrence of "°C"
    for element in soup.find_all(text=True):
        if "°C" in element:
#         if "at 2" in element:
            number = re.search(r"[\d.]+", element.strip())
            

            return float(number.group())  # Return the first occurrence with "°C"
    
    return f"No information containing '°C' found in the section '{section_name}' on {compound_name} page."

# Example usage
compound_name = "1-Octanol"
section_name = "Boiling-Point"   #"Boiling-Point"
temperature_info = get_temperature_info(compound_name, section_name)
print(temperature_info)
```

```python
import requests
import csv

def check_vendor_section(cid):
    """Check if the chemical with the given CID has a vendor section."""
    url = f"https://pubchem.ncbi.nlm.nih.gov/rest/pug_view/data/compound/{cid}/JSON"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        # Look for 'Chemical Vendors' section in record data
        sections = data.get("Record", {}).get("Section", [])
        for section in sections:
            if section.get("TOCHeading") == "Chemical Vendors":
                return 1  # Vendor section found, return 1
    return 0  # No vendor section, return 0

def check_cid_range_and_save(start_cid, end_cid, batch_size=10000):
    """Check CIDs in batches and save results every batch."""
    results = []
    for cid in range(start_cid, end_cid + 1):
        print(f"Checking CID {cid}...")
        vendor_info = check_vendor_section(cid)
        results.append([cid, vendor_info])

        # Save every batch_size records
        if cid % batch_size == 0:
            save_results(results, f"vendor_info_{cid // batch_size}.csv")
            results = []  # Clear results after saving
    
    # Save any remaining records after the loop ends
    if results:
        save_results(results, f"vendor_info_{(end_cid // batch_size) + 1}.csv")

def save_results(data, filename):
    """Save the data to a CSV file."""
    with open(filename, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["CID", "Vendor_Info"])  # Header
        writer.writerows(data)
    print(f"Saved {len(data)} records to {filename}")

# Usage
start_cid = 1
end_cid = 1000000  # Check up to CID 1,000,000
check_cid_range_and_save(start_cid, end_cid)
```
